```python
#Import Statements

import matplotlib.pyplot as plt

import numpy as np #since we are plotting from numpy arrays
```

```python
#Plot a Line
xpoints = np.array([0, 20, 100]) #numpy array with x-coordinates
ypoints = np.array([0, 30, 100]) #numpy array with y-coordinates
plt.plot(xpoints, ypoints) #plot() draws points and by default connects with straight line, first paramter xpoints array, second parameter ypoints array
plt.show() #displays plot
```

```python
#Draw Points
xpoints = np.array([0, 1, 44, 34, 23, 100])
ypoints = np.array([0, 24, 53, 23, 30, 100])
plt.plot(xpoints, ypoints, 'o') #plots points instead of connecting it with a line
plt.show()

ypoints = np.array([3, 8, 1, 10, 5, 7])
plt.plot(ypoints) #when no x-coordinates are given, prints x coordinates are [0,1,2...] by deafult
plt.show()
```

```python
#Formatting
    #[Markers: 'o' circle, '*' star, '.' point, ',' pixel, 'x' X, 'X' x(thick), '+', ...more on w3schools]
    #[Line: '-' solid, ':' dotted, '--' dashed, '-.' dashed/dotted]
    #[Colour: 'r' red, 'b' blue, 'g' green, 'c' cyan, 'm' magenta, 'y' yellow, 'k' black, 'w' white, also hexidecimal + names (ex. hotpink] https://www.w3schools.com/colors/colors_names.asp
plt.plot(ypoints, marker = 'X') #prints a circular mark on each points
plt.show()


plt.plot(ypoints, marker='+', ms = 100, mec = '#4CAF50', mfc ='m') #marker = symbol to mark each point, ms = marker size, mec = marker edge colour, mfc = marker colour
plt.show()
```

```python
#Linestyle
ypoints = np.array([3,8,10,21])
plt.plot(ypoints, linestyle = 'dotted')
plt.show()

plt.plot(ypoints, ls = '--', color = "darkblue", linewidth = '20') #ls shortform of linestyle, linewidth = thickness
plt.show()

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()                      #plotting multiple lines, plt.plot(y1,y2) would take y1 as x-values

x1 = np.array([0, 2, 4, 6])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1, y1, x2, y2)
plt.show()                      #plotting multiple lines with x-values
```

```python
#Labels
font1 = {'family':'serif','color':'blue','size':20}     #setting Font Properties for a Label
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Aug 1st Data", fontdict=font1, loc='left')   #setting titles with optional font and optional loc can be left/right/center
plt.xlabel("Time", fontdict=font2)                      #setting axis label with optional font
plt.ylabel("Displacement", fontdict = font2, loc = 'bottom') #loc for y can be bottom/center/top
x1 = np.array([0, 2, 4, 6])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1,y1,x2,y2)
plt.show()
```
